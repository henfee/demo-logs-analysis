#Uber GPS Logs from Taxis in SFO

This notebook works against a set of anonimized taxi location logs provided by Uber to determine hotspots in the city of SFO. These logs have been uploaded into BigQuery, and are accessed using SQL in the notebook.

You can read more about the dataset at http://www.infochimps.com/datasets/uber-anonymized-gps-logs.

```python
import samples.cloud as cloud
from samples.mapping import HeatMap
```

```python
logs_table = cloud.BigQuery.table('[uberlogs.logs]')
```

##Exploration

Lets start by looking at the schema of the logs, and issuing a SQL query to get a sampling of the logs.

```python
logs_table.schema()
```

```python
%%bigQuery
SELECT trip, timestamp, lat, lon
FROM $logs_table
WHERE trip = 1
ORDER BY timestamp
LIMIT 5
```

##Spatial Aggregation

As you'll notice, the GPS signals are effectively spaced 4ms apart, and have discrete latitude/longitude coordinates. As such they don't group well... we need to come up with a key to group the individual entries.

Location coordinates can be geo-hashed. For the purposes of this notebook, we'll instead choose a simpler methodology - simply make our coordinates have 3-decimal place precision, and then concatenating them to produce a key. All of this can be easily done in the SQL, as shown below.

```python
%%bigQuery geohashes
SELECT trip, lat, lon, CONCAT(STRING(FLOOR(lat*1000+0.5)/1000), ':', STRING(FLOOR(lon*1000+0.5)/1000)) AS geoid
FROM $logs_table
```

Lets pick a few samples of results to ensure we have expected values for our computed geoid column.

```python
%%bigQuery
SELECT trip, lat, lon, geoid
FROM $geohashes
LIMIT 5
```

##Counting

Now that we have a key to group over, we can simply perform grouping, and aggregating the trip ids to produce a trip count for each geoid.

Also, for the purposes of the demo, since we're building an interactive map, lets pick a small sampling of our data. In particular, we're using BigQuery's ability to sample data and picking a .1% sample of the data.

```python
%%bigQuery activity
SELECT FIRST(lat) AS lat, FIRST(lon) AS lon, COUNT(trip) AS trips
FROM $geohashes
WHERE ABS(HASH(trip)) % 1000 == 1
GROUP BY geoid
ORDER BY trips DESC
```

```python
df = activity.dataFrame()
df[:10]
```

```python
map = HeatMap()
map.render('activity_map', activity.list())
```

##Conclusion

This simple agggregation over GPS data can be applied to many logs scenarios that include location attributes to derive spatial understanding of the data, or the sources generating that data.
